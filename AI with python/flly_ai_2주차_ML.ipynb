# 딥러닝
기계가 학습의 한 분야인 인공신경망에 기반해서 많은 양의 데이터를 학습해 뛰어난 성능을 이끌어내는 연구분야

# 인공신경망 개요
- 기계학습과 전통적인 프로그래밍 차이(직접 입력 데이터를 통한 처리 <-> 입력 데이터를 넣으면 스스로 프로그래밍)

# NN vs DNN
ANN기법(NN)의 여러문제가 해결되면서 모델 내 은닉층을 많이 늘려서 학습의 결과를 향상 -> DNN(Deep Neural Network) </br>
- hidden layer를 어떻게 쌓는냐에 따라 결과가 달라짐
- DNN은 은닉층을 2개 이상 지닌 학습 방법을 의미

# ANN
인공신경망이라고 불리는 ANN은 사람의 신경망 원리와 구조를 모방하여 만든 기계학습 알고리즘
### 구성
- 다수의 입력 데이터를 받는 입력층
- 데이터의 출력을 담당하는 출력층
- 입력층과 출력층 사이에 존재하는 레이어들은 입력층
- 모델 구성
  - 히든 레이어들의 갯수와 노드의 개수를 구성하는 것
  - output값을 잘 예측하는 것을 목표로 함
  - 은닉층에서는 활성화함수를 사용(활성화함수? 최적의 Weight와 Bias를 찾아내는 역할)
### 문제점
- 학습과정에서 파라미터의 최적밧을 찾기 어렵다 <- 블랙박스 모델
- Overfitting에 따른 문제
- 학습시간이 너무 느리다. (fully conneted이기 때문임)
### 활성화 함수
어떻게 데이터들의 가중치 등을 줄지 선택할 수 있음
- Sigmoid function : 로지스틱 회귀모형의 대표로 0~1사이값으로 변환, (0 -> 1)
- Tanh function : 스케일이 -1 ~ 1사이 값으로 변환
- ReLU function : 입력값이 0을 넘으면 그대로 출력, 0 이하는 0으로 출력(0 -> n)

### 가중치 업데이트
- 오차 역전파법
  - 학습하는 메커니즘을 설명하는 알고리즘
  - 모형의 오차를 최소화하기 위해 출력층부터 입력층까지 역의 방향으로 오차를 전파하며 가중치와 편향을 조정
  - 격차를 줄이기 위한 방법이라고 생각하면 편함
### 손실함수
- 평균제곱오차(MSE)
  - 수치를 예측할때 사용되는 손실함수로 실제값(y)과 예측값( )의 차이를 제곱한 값의 평균.
  - 손실함수의 값이 최소화 될수록 최적화도니 모형, MSE를 최소화하는 방향으로 모형이 학습
- 교차 엔트로피 오차(CEE)
  - 종속변수가 범주형일때 주로 사용
  - y는 원핫인코딩된 범주, yi는 특정범주에 속할 확률
  - CEE를 최소화하는 방향으로 모형이 학습

## 회귀분석
### 지도학습
- 명확한 결정값이 주어진 데이터를 학습(데이터 레이블 필요)
  - 분류
  - 회귀
  - 시각/음성감지/인지

### 비지도학습
- 결정값이 주어지지 않은 데이터를 학습
  - 군집화(클러스터링)
  - 차원 숙소 : 데이터가 가지고 있는 함축적 의미들을 추출
### 강화학습
- 피드백을 통해 머신 러닝 알고리즘을 학습
- "좋은" 출력과 "나쁜" 출력이 있으며, 시간이 지남에 따라 머신 러닝 알고리즘은 나쁜 출력을 피하는 방법을 학습

```python
#다항회귀 코드

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

auto_mpg = pd.read_csv('/content/auto-mpg.csv')

# 1. 데이터 준비
X_train = auto_mpg.loc[:, ['displacement']].values  # 입력 데이터
y_train = auto_mpg.loc[:, ['weight']].values  # 타겟 값
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")

# 2. 다항식 변환
degree = 2  # 다항식 차수 설정
poly_features = PolynomialFeatures(degree=degree, include_bias=False)
X_poly = poly_features.fit_transform(X_train)  # 다항식 변환

# 3. 다항 회귀 모델 학습
model = LinearRegression()
model.fit(X_poly, y_train)

# 4. 예측값 계산 (곡선 그리기용)
X_range = np.linspace(X_train.min(), X_train.max(), 100).reshape(-1, 1)  # X 범위 설정
X_range_poly = poly_features.transform(X_range)  # 다항식 변환
y_pred = model.predict(X_range_poly)  # 예측값 계산

# 5. 시각화
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Original Data')  # 원본 데이터
plt.plot(X_range, y_pred, color='red', label=f'Polynomial Regression (degree={degree})')  # 회귀 곡선
plt.title("Polynomial Regression Curve for Displacement vs Weight", fontsize=16)
plt.xlabel("Displacement", fontsize=12)
plt.ylabel("Weight", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Confusion matrix - 예측값
- 정확도(accuracy)
  - 전체의 경우 중에 맞은 경우에 대한 확률, (TN+TP) / (전체)
- 정밀도 (precision)
  - 맞다고 예측한 애들 중에 실제로 맞은것의 비율, TP / (FP+TP)
- 재현율 (recall)
  - 실제로 맞은 사람들 중에 맞았다고 예측한 애들의 비율, TP / (FN+TP)
- F1 스코어
  - 정밀도와 재현율은 서로 트레이드오프 관계, 둘 다 높은 점수를 맞아야 좋은 모델이라 할 수 있음
  - F1 스코어는 정밀도와 재현율의 조화 평균, 2 / [(1/recall) + (1/precision)]
- ROC, AUC
  - 맞출 때마다 TPR(수직) 방향으로 한 칸씩 올라갑니다. 그리고 틀리면 FPR(수평) 방향으로 이동

- LBP(Local Binary Patterns, 지역 이진 패턴)는 텍스처 분석에 자주 사용되는 간단하면서도 강력한 특징 기술
- 이 기법은 이미지의 지역적인 텍스처 패턴을 기반으로 작동하며, 텍스처의 기본적인 구조를 이진 숫자로 표현

```python
from skimage import feature, color, io
import numpy as np

def extract_features(image_path):
    image = io.imread(image_path)
    image_gray = color.rgb2gray(image)
    # LBP 특징 추출
    lbp = feature.local_binary_pattern(image_gray, P=24, R=3, method="uniform")
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, 27), range=(0, 26))
    lbp_hist = lbp_hist.astype("float")
    lbp_hist /= (lbp_hist.sum() + 1e-6)  # 정규화
    return lbp_hist

# 폴더에서 이미지를 불러와 특징 추출
folder_paths = ['/content/shapes/circles', '/content/shapes/squares', '/content/shapes/triangles']
labels = []
features = []

for label, folder_path in enumerate(folder_paths):
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.png'):
            file_path = os.path.join(folder_path, file_name)
            img_features = extract_features(file_path)
            features.append(img_features)
            labels.append(label)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# 특징 배열과 레이블 배열을 준비
X = np.array(features)
y = np.array(labels)

# 데이터를 훈련 세트와 테스트 세트로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 결정 트리 모델 생성 및 훈련
tree_model = DecisionTreeClassifier(max_depth=5, random_state=42)
tree_model.fit(X_train, y_train)
```

predict() 함수는 주어진 입력 데이터에 대해 모델이 학습된 패턴을 기반으로 예측을 수행한다. 이 예측은 분류(classification) 작업에서는 클래스 레이블을, 회귀(regression) 작업에서는 연속적인 값(output)을 반환

```python
from sklearn.metrics import accuracy_score

# 테스트 데이터에 대한 예측 수행
y_pred = tree_model.predict(X_test)

# 정확도 계산
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
