## Numpy

- 다차원배열 생성
- 배열을 속성
- 배열을 자료형
- 배열의 형태와 축
- → 이 리스트 형태를 데이터프레임(엑셀)으로 바꿔준 것이 판다스
- 리스트에서 데이터를 입력 받고 여기서 슬라이스 하는 것이 핵심
- 반복문 없이 데이터 배열의 처리를 지원하고 빠르고 편리

### Numpy 배열의 차원

- 차원 : 관측하고자 하는 데이터의 속성 수 또는 항목의 수
- 스칼라 : 0차원 (배열에서 값을 표현하는 가장 기본 단위로 하나의 실수를 저장)
- 백터 : 1차원 배열, 스칼라 여러 개를 나열한 튜플
- 행렬 : 2차원 배열, 1차원 배열을 여러 . 개묶은 배열
- 텐서 : 벡터의 집합, 3차원 이상의 배열
- 랭크 : 차원의 수
- 모양 : 배열의 차원과 크기
- ex. 행이 3개, 열이 4개 2차원 배열 → 랭크 = 2, 모양 = (3,4)

### ndarry

- 넘파이의 다차원 배열 객체를 ndarry
    - 자료형이 모두 같은 데이터를 담은 다차원 배열
    - 정수 또는 실수(부동 소수점 수)를 저장 (메모리 할당이 선점형이 아님)
    - 배열 데이터에도 순서가 있으므로 인덱싱과 슬라이싱이 가능(행을 기준? 열을 기준?으로 자를건지 슬라이싱)

- ex. arr = np.array([1,2,3,4] // 1차원
- ex. print(arr)
- ex. print(type(arr))
- ex. arr = np.array([[1,2,3],[4,5,6]]) // 2차원

### 넘파이 배열의 인덱싱과 축

- 배열을 정수나 다른 배열, 논리값, 음이 아닌 정수 튜플로 인덱싱
- 3차원 면, 행, 열 순으로 인덱스

### 다차원 배열 생성

- shape : 배열의 형태를 튜플 자료형으로 표현
- ndim : 몇차원의 배열인지 나타냄
- size : 원소의 개수를 나타냄
- type : 배열원소의 자료형을 나타냄
- itemsize : 배열 원소의 크기를 바이트 단위로 표시 → 메모리 크기

## Evibigdata 데이터를 사용한 실습

````python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import numpy as np

data = pd.read_csv('/content/sample_data/dust_data.csv')

data['issuedate'] = pd.to_datetime(data['issuedate'])

data['month'] = data['issuedate'].dt.to_period('M')
monthly_alerts = data.groupby(['districtname', 'month']).size().unstack(fill_value=0)

plt.rc('font', family='NanumBarunGothic')

districts = monthly_alerts.index
colors = cm.rainbow(np.linspace(0, 1, len(districts)))
district_color_map = dict(zip(districts, colors))

# 그래프
plt.figure(figsize=(14, 8))
for district in districts:
    x_values = [str(month) for month in monthly_alerts.columns]
    plt.plot(x_values, monthly_alerts.loc[district], marker='o', color=district_color_map[district], label=district)

plt.title('2019 지역별 미세먼지 경보 발령 빈도 (월별)')
plt.xlabel('월')
plt.ylabel('경보 발령 횟수')
plt.xticks(rotation=45)
plt.legend(title='지역', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
````
